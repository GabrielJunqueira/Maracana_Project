```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#normal day, to compare
df = pd.read_csv('../../../API scripts/01.Get the Trips Hourly/Hourly(21-9-22)/hourly(21-9-22).csv')
df = df[df["origin_id"]!='STADIUM AREA, (182)']
```

```python
df = df[["total_trips" , "Arrival Hour"]]
df = df.groupby('Arrival Hour').sum()
```

### Saving

```python
df.to_csv('trips(21-9-22).csv')
```

```python
# Create a bar plot
df.plot.bar(figsize=(18,6))
```

## Comparing with match day

```python
match_day_trips = pd.read_csv('../../../data/data from API/match_day/trips(20-4).csv')
match_day_trips = match_day_trips.drop(columns=["index", "Unnamed: 0"], axis=1)
```

```python
# Data for first bar plot
data1 = df['total_trips']

# Data for second bar plot
data2 = match_day_trips['total_trips']

# Create figure and axes
fig, ax = plt.subplots(figsize=(18,12))

# Plot first bar plot
ax.bar(range(len(data1)), data1, width=0.4, align='edge', label='Normal Wednesday')

# Plot second bar plot
ax.bar(np.arange(len(data2))+0.4, data2, width=0.4, align='edge', label='Match Day')

ax.legend()
plt.show()
```
