```python
#Function to normalize the "total_trips" column
def normalize(df):
    for key, value in df["total_trips"].iteritems():
        if value == '<10':
            df.at[key, 'total_trips'] = '5'
    
    df['total_trips'] = df['total_trips'].astype(float)
    print(df['total_trips'].sum())
```

```python
normalize(normal_day)
normalize(match_day)
normalize(flu_day)
```
