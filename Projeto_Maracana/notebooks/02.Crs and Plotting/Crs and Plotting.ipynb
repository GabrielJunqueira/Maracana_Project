```python
import pandas as pd
import geopandas as gpd
%matplotlib inline
import osmnx as ox
import networkx as nx
```

```python
streets =  gpd.read_file('../../data/StreetNetwork/G2_edges.shp')
```

```python
buildings = gpd.read_file('../../data/Buildings/buildings.geojson')
```

```python
maracana = gpd.read_file('../../data/Maracana/maracana.geojson')
```

```python
location_point = (-22.911667, -43.230278)
dist = 2300
```

# Crs Maracana

```python
maracana = maracana.to_crs("EPSG:4326")
```

### Saving

```python
import os
dir = '../../data/Maracana'
```

```python
maracana.to_file(dir + '/maracana.geojson')
```

# Plotting

```python
streets_ox= ox.graph_from_point(location_point, network_type='drive', dist=dist)
```

```python
fig, ax = ox.plot_graph(streets_ox, node_size=0, edge_linewidth=0.5, show=False, close=False, figsize=(22,12))
streets.plot(ax=ax, color="White", figsize=(22,12))
maracana.plot(ax=ax, color="Red", figsize=(22,12))
buildings.plot(ax=ax, color="White", figsize=(22,12))
```
